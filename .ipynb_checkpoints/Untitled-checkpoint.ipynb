## Creating a Custom fits file

```python
from astropy.io import fits
```

```python
hdulist = fits.HDUList()
```

```python
hdu1 = fits.PrimaryHDU()
```

```python
hdu1.data = M31_Pixels
```

```python
hdu1.header = M31_header
```

```python
hdulist.append(hdu1)
```

```python
hdulist.writeto('Dummy_FITS_file2.fits' , overwrite=True )
```
